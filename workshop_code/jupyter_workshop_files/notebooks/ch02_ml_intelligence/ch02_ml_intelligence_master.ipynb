### 1️⃣ CSV Loading (Beginner)
การดึงข้อมูล Agriculture Data จากไฟล์ CSV

```python
import pandas as pd
df = pd.read_csv('../../datasets/crop_data_v2.csv')
df.head()
```

### 2️⃣ Data Cleaning (Intermediate)
จัดการข้อมูลเซนเซอร์ที่หายไปด้วยค่าเฉลี่ย

```python
df_clean = df.fillna(df.mean())
print(f"ทำความสะอาดข้อมูลเรียบร้อย: {len(df_clean)} แถว")
```

### 3️⃣ Random Forest Prediction (Advanced)
ทายผลผลิตจากข้อมูลฟาร์ม (READY vs LOW_YIELD)

```python
from sklearn.ensemble import RandomForestClassifier
X = df_clean.drop(['label', 'yield_weight'], axis=1)
y = df_clean['label']

model = RandomForestClassifier()
model.fit(X, y)
print(f"🔮 ทายผลผลิตจากเซนเซอร์ใหม่: {model.predict([[35, 75, 40, 15, 12]])}")
```
