### 3️⃣ Decision Logic (Intermediate)
การใช้ If-Else เพื่อตัดสินใจเปิด-ปิดพัดลม

```python
temp = 36
if temp > 35: 
    print("🔥 เปิดพัดลม")
else:
    print("✅ อากาศปกติ")
```

### 4️⃣ Smart Loop (Intermediate)
การวนลูปตรวจสอบเซนเซอร์ทั้งฟาร์ม

```python
sensors = [32, 35, 38, 30, 31]
for s in sensors:
    if s > 35: print(f"🚨 จุดตรวจเจอความร้อน: {s}")
```
