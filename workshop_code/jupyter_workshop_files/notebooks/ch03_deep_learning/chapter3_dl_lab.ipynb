## 🕒 Lab Activity 1

```python
# ReLU คือคนคัดกรอง: ถ้าเป็นค่าลบให้เป็น 0 ถ้าเป็นบวกให้ผ่านได้
sample_values = [-5, 12, -2, 8]
relu_output = [max(0, x) for x in sample_values]
print(f"ผลลัพธ์จาก ReLU สำหรับ {sample_values} คือ {relu_output}")
```
