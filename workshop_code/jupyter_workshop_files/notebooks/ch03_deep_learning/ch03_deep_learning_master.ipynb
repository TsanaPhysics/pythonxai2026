### 1️⃣ Activation Functions (Beginner)
การทำงานของ ReLU และ Sigmoid (การพิกัดค่าความน่าจะเป็น)

```python
import numpy as np
def relu(x): return np.maximum(0, x)
print(f"ReLU Input -10: {relu(-10)} | ReLU Input 10: {relu(10)}")
```

### 2️⃣ Artificial Neuron (Intermediate)
จำลองการทำงานของเซลล์ประสาท 1 เซลล์ (Weight & Bias)

```python
w, b, val = 0.8, -2.0, 10.0
z = (val * w) + b
print(f"📍 Linear Output: {z} -> Activated (ReLU): {max(0, z)}")
```

### 3️⃣ Sequential Model Logic (Advanced)
การเชื่อมต่อ Layer สู่ Layer เพื่อประมวลผลข้อมูลจากหลายปัจจัย

```python
inputs = [1, 0.5]
weights = [0.8, 0.2]
hidden = max(0, sum([i*w for i,w in zip(inputs, weights)]))
print(f"🧠 Hidden Layer Active Node: {hidden}")
```
