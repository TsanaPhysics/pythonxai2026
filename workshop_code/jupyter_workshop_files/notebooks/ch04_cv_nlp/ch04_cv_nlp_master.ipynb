### 1️⃣ Pixel Access & RGB (Beginner)
วิเคราะห์เฉดสีพื้นฐานเพื่อตรวจสุขภาพพืช

```python
color = (34, 139, 34) # เขียวปกติ
if color[1] > color[0]: print("✅ ใบพืชปกติดี (High Green)")
```

### 2️⃣ HSV Thresholding Logic (Advanced)
การใช้ช่วงเฉดสีเพื่อตรวจหารอยโรคสีน้ำตาล

```python
low_brown = [10, 100, 20]
pixel_hsv = (15, 150, 120)
if low_brown[0] <= pixel_hsv[0] <= 30: print("🚨 พบเป้าหมาย: รอยโรค (Brown Spot)")
```

### 3️⃣ Keyword Spotting & Intent (Intermediate)
การแปลงเสียงพูดเป็นคำสั่งทำงานอัตโนมัติ

```python
voice = "รดน้ำต้นไม้"
if "รดน้ำ" in voice: print("🤖 สั่งการ: WATER_ON")
```

### 4️⃣ Sentiment Analysis (Advanced)
การวิเคราะห์อารมณ์ผู้พูด

```python
text = "ระบบพังอีกแล้ว แย่มาก"
bad_words = ["พัง", "แย่"]
count = sum([1 for w in bad_words if w in text])
if count > 0: print("😟 พบอารมณ์เชิงลบ (Negative)")
```
