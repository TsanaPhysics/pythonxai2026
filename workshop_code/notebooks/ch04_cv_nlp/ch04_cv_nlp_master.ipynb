### 1️⃣ Computer Vision: Pixel Access
การเข้าถึงรหัสสีและพิกัดในภาพถ่ายพืช

```python
# ex01_pixel_access.py
x, y = 10, 10
color = (34, 139, 34) # R, G, B
print(f"📍 พิกัด: {x}, {y} | 🎨 สีที่พบ: R={color[0]}, G={color[1]}, B={color[2]}")

if color[1] > color[0] and color[1] > color[2]:
    print("✅ ระบบสรุป: นี่คือส่วนของ 'ใบพืช' ที่ปกติดี")
else:
    print("⚠️ ระบบสรุป: พบเฉดสีผิดปกติ!")
```

### 2️⃣ CV: HSV Masking for Disease Detection
การใช้โมเดลสี HSV เพื่อคัดกรองรอยโรค (Brown Spot)

```python
# ex02_hsv_mask.py & ex08_hsv_masking.py
low_hsv = (10, 50, 20)
high_hsv = (30, 255, 200)
current_hsv = (15, 150, 100)

print(f"🎯 เป้าหมาย: ตรวจหาจุดสีในช่วง {low_hsv} ถึง {high_hsv}")
if all(l <= c <= h for l, c, h in zip(low_hsv, current_hsv, high_hsv)):
    print("🚨 พบรอยโรค! พิกัดนี้คือเฉดสีน้ำตาล (Sick Area)")
else:
    print("✅ พิกัดนี้ปกติ (Healthy Area)")
```

### 3️⃣ NLP: Sentiment Analysis
การวิเคราะห์อารมณ์ความรู้สึกของผู้ใช้งาน (Sentiment Score)

```python
# ex04_sentiment_basic.py
good_words = ["ดี", "ชอบ", "สุดยอด", "ขอบคุณ"]
bad_words = ["พัง", "ไม่ดี", "แย่", "หยุด"]
user_text = "ขอบคุณมากครับ ระบบทำงานได้ดีมาก"
score = sum(1 for w in user_text.split() if w in good_words) - sum(1 for w in user_text.split() if w in bad_words)

if score > 0: print("😊 AI สัมผัสถึง: อารมณ์เชิงบวก (Positive)")
elif score < 0: print("😟 AI สัมผัสถึง: อารมณ์เชิงลบ (Negative)")
else: print("😐 AI สัมผัสถึง: กลางๆ (Neutral)")
```

### 4️⃣ NLP: Intent Mapping
การแปลงคำพูดให้เป็นคำสั่งงาน (Intent Classification)

```python
# ex03_intent_map.py & ex09_intent_mapping.py
intents = {
    "รดน้ำพืชหน่อย": "WATERING_ON",
    "อากาศเป็นยังไง": "WEATHER_REPORT",
    "ช่วยดูแลทุเรียน": "MONITORING_START"
}
user_input = "อากาศเป็นยังไง"
action = intents.get(user_input, "UNKNOWN_COMMAND")
print(f"👂 เสียงผู้ใช้: {user_input} -> 🤖 คำสั่ง AI: {action}")
```

### 5️⃣ Agri-Bot Liaison (Chatbot Logic)
ระบบตอบกลับอัตโนมัติเบื้องต้น

```python
# ex05_bot_liaison.py
responses = {
    "สวัสดี": "สวัสดีครับ! ผม Agri-Bot มีอะไรให้ช่วยไหม?",
    "สบายดีไหม": "ผมเป็น AI ครับ สบายดีและพร้อมรดน้ำต้นไม้เสมอ!",
    "ขอบคุณ": "ยินดีเสมอครับ ขอให้ได้ผลผลิตที่ดีนะครับ"
}
def get_response(text):
    for key in responses:
        if key in text: return responses[key]
    return "ไม่เข้าใจครับ ผมกำลังเรียนรู้ประโยคใหม่ๆ อยู่"

print(f"🤖 Bot: {get_response('สวัสดี')}")
print(f"🤖 Bot: {get_response('ขอบคุณ')}")
```
