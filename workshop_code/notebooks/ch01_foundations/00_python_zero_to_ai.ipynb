## 🌡️ 1. ตัวแปรและประเภทข้อมูล (Variables & Data Types)
ตัวแปรคือ "กล่อง" ที่เราใช้เก็บค่าจากเซ็นเซอร์หรือข้อมูลพืช

```python
# ลองรันโค้ดตรงนี้ (Shift + Enter)
crop = "ข้าวหอมมะลิ"    # ข้อความ (String)
temp = 32.5            # ตัวเลขทศนิยม (Float)
is_auto = True         # ค่าความจริง (Boolean)

print(f"พืช: {crop} | อุณหภูมิ: {temp} | ระบบอัตโนมัติ: {is_auto}")
```

--- 
## 📦 2. การเก็บข้อมูลเป็นชุด (Collections)
AI ต้องจัดการข้อมูลจำนวนมาก เราจึงต้องรู้จักวิธีเก็บข้อมูลแบบหลายค่า

```python
# List: เก็บรายการพืชในแปลง
fruits = ["เมลอน", "มะเขือเทศ", "ผักสลัด"]
print(f"รายการที่ 1: {fruits[0]}") # เริ่มนับที่ 0 เสมอ!

# Dictionary: เก็บข้อมูลเชิงลึก (Key-Value)
sensor = {
    "id": "SN-001",
    "value": 45,
    "unit": "%"
}
print(f"เซ็นเซอร์ {sensor['id']} อ่านค่าได้ {sensor['value']} {sensor['unit']}")
```

--- 
## 🧠 3. ตรรกะและการตัดสินใจ (Logic & Decision Making)
นี่คือหัวใจของ Smart Farm: "ถ้า... แล้ว..."

```python
soil_moisture = 35 # ค่าความชื้นปัจจุบัน

if soil_moisture < 40:
    print("🚨 ความชื้นต่ำ! เปิดปั๊มน้ำอัตโนมัติ...")
elif soil_moisture > 80:
    print("⚠️ ความชื้นสูงเกินไป! ปิดระบบรดน้ำ")
else:
    print("✅ สภาวะปกติ")
```

--- 
## 🔄 4. การทำงานซ้ำ (Loops)
ใช้ประมวลผลข้อมูลย้อนหลัง หรือตรวจสอบพืชทุกต้นในแปลง

```python
temps = [28, 30, 35, 33, 29]

for t in temps:
    if t > 34:
        print(f"อุณภูมิ {t} -> ☀️ ร้อนจัด")
    else:
        print(f"อุณภูมิ {t} -> 🌤️ ปลอดโปร่ง")
```

--- 
## 🛠️ 5. ฟังก์ชันและเครื่องมือ (Functions & Libraries)
สร้างเครื่องมือส่วนตัว และนำเข้าความสามารถจากภายนอก (Pandas, OpenCV)

```python
def calculate_yield(area, rate):
    return area * rate

result = calculate_yield(10, 500) # พื้นที่ 10 ไร่, อัตรา 500 กก./ไร่
print(f"ผลผลิตที่คาดการณ์: {result} กิโลกรัม")

# การนำเข้าไลบรารี
import math
print(f"ค่าคงที่ PI สำหรับคำนวณพื้นที่วงกลม: {math.pi}")
```
