### 1️⃣ Activation Functions: ReLU Logic
ทำความเข้าใจการทำงานของ Rectified Linear Unit (ReLU)

```python
# ex01_relu_demo.py & ex07_relu_activation.py
import numpy as np

def relu(x):
    return np.maximum(0, x)

sample_values = [-10, -5, 0, 8, 12]
outputs = [relu(x) for x in sample_values]

print("--- [Deep Learning: ReLU Activation] ---")
for x, y in zip(sample_values, outputs):
    print(f"Input: {x:3} -> Output: {y}")
```

### 2️⃣ Activation Visualization
เปรียบเทียบ ReLU และ Sigmoid ผ่านกราฟ

```python
# ex02_activation_vis.py
import matplotlib.pyplot as plt

def sigmoid(x): return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y_relu = relu(x)
y_sig = sigmoid(x)

plt.plot(x, y_relu, label='ReLU', color='orange')
plt.plot(x, y_sig, label='Sigmoid', color='cyan')
plt.title('Activation Functions: Comparing Relay vs Probabilities')
plt.grid(True)
plt.legend()
plt.show()
```

### 3️⃣ Brain Node (The Neuron)
จำลองการทำงานของ Neuron (Weight & Bias)

```python
# ex03_brain_node.py
def neuron(input_val, weight, bias):
    # ผลบวกเชิงเส้น (Linear Combination)
    z = (input_val * weight) + bias
    # การเปิดใช้งาน (Activation)
    a = max(0, z) # ReLU
    return a

w = 0.8
b = -2.0
val = 10.0
result = neuron(val, w, b)
print(f"📍 Input: {val} * Weight: {w} + Bias: {b} = Result: {result}")
```

### 4️⃣ Sequential Model Simulation
จำลองโครงสร้าง Sequential Model (Hidden Layer -> Output)

```python
# ex04_sequential_model.py
def layer(inputs, weights, biases):
    return [max(0, i * w + b) for i, w, b in zip(inputs, weights, biases)]

input_data = [1.0, 0.5, -1.2]
hidden_w = [0.8, 0.2, 0.1]
hidden_b = [0, -0.5, 0.2]
h1 = layer(input_data, hidden_w, hidden_b)
print(f"🧠 Hidden Layer Results: {h1}")

output_w = [0.9, 0.9, 0.9]
output_b = [0, 0, 0]
final = layer(h1, output_w, output_b)
print(f"🏁 Final Prediction (Sum): {sum(final)}")
```

### 5️⃣ Overfitting Diagnosis
การวินิจฉัยปัญหาการจำข้อสอบของ AI (Overfitting)

```python
# ex05_overfitting_check.py
train_acc = 99.8 # แม่นยำในห้องเรียน
test_acc = 65.2  # แม่นยำเมื่อเจอของจริง
print(f"📉 Train Accuracy: {train_acc}% | Test Accuracy: {test_acc}%")

if train_acc - test_acc > 20: 
    print("⚠️ สถานะ: พบปัญหา Overfitting!")
else:
    print("✅ สถานะ: โมเดลมีคุณสมบัติที่ดี (Good Generalization)")
```
