### 3️⃣ Smart Move (Autonomous Logic)
ตรรกะการเคลื่อนที่เข้าหาเป้าหมายโดยอัตโนมัติ

```python
%%writefile current_game.py
import pgzrun
WIDTH, HEIGHT = 800, 600
harvester = Actor('harvester')
target_x, target_y = 600, 100
def draw():
    screen.clear()
    screen.fill((34, 139, 34))
    harvester.draw()
    screen.draw.circle((target_x, target_y), 5, "red")
def update():
    if harvester.x < target_x: harvester.x += 2
    elif harvester.x > target_x: harvester.x -= 2
    if harvester.y < target_y: harvester.y += 2
    elif harvester.y > target_y: harvester.y -= 2
pgzrun.go()
```

### 4️⃣ Collision Sensor
การตรวจจับการชน (Collision Detection) ระหว่าง Harvester และพืช

```python
%%writefile current_game.py
import pgzrun
WIDTH, HEIGHT = 800, 600
harvester = Actor('harvester')
harvester.pos = 100, 100
plant = Actor('plant')
plant.pos = 400, 300
def draw():
    screen.clear()
    screen.fill((34, 139, 34))
    harvester.draw()
    plant.draw()
    if harvester.colliderect(plant):
        screen.draw.text("DETECTED: AGRI-OBJECT FOUND!", (20, 20), color="yellow")
def update():
    if keyboard.right: harvester.x += 5
    if keyboard.left: harvester.x -= 5
    if keyboard.up: harvester.y -= 5
    if keyboard.down: harvester.y += 5
pgzrun.go()
```

### 5️⃣ AI Harvester Simulation (Full Autonomous)
การสร้าง Agent สำหรับดูแลรักษาพืชที่เป็นโรคโดยอัตโนมัติ

```python
%%writefile current_game.py
import pgzrun
import random
WIDTH, HEIGHT = 800, 600
harvester = Actor('harvester', pos=(400, 300))
plants = []
for i in range(3):
    p = Actor('plant', pos=(random.randint(100, 700), random.randint(100, 500)))
    p.sick = random.choice([True, False])
    plants.append(p)
def draw():
    screen.clear()
    screen.fill((34, 139, 34))
    harvester.draw()
    for p in plants:
        p.draw()
        if p.sick: screen.draw.circle(p.pos, 15, "red")
def update():
    for p in plants:
        if p.sick:
            if harvester.x < p.x: harvester.x += 2
            elif harvester.x > p.x: harvester.x -= 2
            if harvester.y < p.y: harvester.y += 2
            elif harvester.y > p.y: harvester.y -= 2
            if harvester.colliderect(p):
                p.sick = False
                print("✨ Plant Cured!")
pgzrun.go()
```
