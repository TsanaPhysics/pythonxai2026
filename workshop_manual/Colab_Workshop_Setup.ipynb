# 🚀 การเตรียมข้อมูล Workshop สำหรับ Google Colab

เพื่อให้สามารถรันโมเดล AI และ Computer Vision ในระบบ Cloud ของ Google ได้โดยตรง

**คำแนะนำขั้นตอน:**
1. กดปุ่ม **[▶ Run]** ที่บล็อกโค้ดด้านล่าง
2. จะมีปุ่ม **[Choose Files]** ปรากฏขึ้นมา
3. ให้เลือกไฟล์ `jupyter_workshop_data.zip` จากเครื่องของคุณ
4. ระบบจะทำการอัปโหลดและแตกไฟล์ให้โดยอัตโนมัติ

```python
from google.colab import files
import zipfile
import os

print("📥 กรุณาเลือกไฟล์ zip ข้อมูลที่ดาวน์โหลดมา:")
uploaded = files.upload()

for fn in uploaded.keys():
    if fn.endswith('.zip'):
        print(f'⏳ กำลังแตกไฟล์ "{fn}"...')
        with zipfile.ZipFile(fn, 'r') as zip_ref:
            zip_ref.extractall('.')
        print("\n✅ สำเร็จ! พบไฟล์ข้อมูลดังต่อไปนี้พร้อมใช้งาน:")
        if os.path.exists('data'):
            for item in os.listdir('data'):
                print('  -', item)
        else:
            print("ไม่พบโฟลเดอร์ data กรุณาตรวจสอบโครงสร้าง Zip")
```
